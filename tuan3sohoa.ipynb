```python
import numpy as np
import pandas as pd
data = {
'ID' : [101,102,103,104,105],
'Name' : ['Quynh Nhu','Hoang Hai',None,'Phuong Tuan','Thien An'],
'Age' : [26,18,20,None,19],
'Salary' : [500,350,650,1000,None]
}
df = pd.DataFrame(data)
df

```

```python
df = pd.DataFrame(data)
df
df['Age'] = df['Age'].fillna(df['Age'].mean())
print(df)
```

```python
df = pd.DataFrame(data)
df
df.dropna(inplace=True)
print(df)
```

```python

# Tạo DataFrame
df = pd.DataFrame(data)

# Điền giá trị thiếu trong cột 'Name' bằng 'unknown'
df['Name'].fillna('unknown', inplace=True)

# Hiển thị DataFrame
print(df)
```

```python
df1=pd.DataFrame({
    "ID":[1,2,3],
    "Score A":[70,90,85]
})
df2=pd.DataFrame({
    "ID":[3,4,5],
    'Score_B':[62,91,75]
})
#Thực hiện merge cột ID(Inner join, left join, outer join
pd.merge(df1,df2,on="ID",how="inner")
#Nối DataFrame theo chiều dọc
pd.concat([df1,df2],axis=0)#axis: chiều dọc, concat: nối 2 data lại với nhau
#Gộp df1 và df2 để điền giá trị thiếu
df1.set_index("ID").combine_first(df2.set_index("ID")).reset_index()
```

```python
#Cho dữ liệu sau:
data1=pd.DataFrame({
    "ID":range(1,1000000),
    "Value": np.random.randint(1,100,999999)
})
#Dùng .astype để tối ưu hóa bộ nhớ


# Chuyển đổi kiểu dữ liệu để tối ưu
data1["ID"] = data1["ID"].astype(np.int32)   # ID là số nguyên dương, có thể dùng uint32 thay vì int64
data1["Value"] = data1["Value"].astype(np.int8)  # Value trong khoảng 1-100, dùng uint8 là đủ

#Tìm 5 giá trị phổ biến trong cột value
data1["Value"].value_counts().head(5)

#Sử dụng query để lọc dữ liệu nhanh hơn df[df["Value"]>90]
data1.query("Value > 90")
```

```python
import matplotlib.pyplot as plt
import numpy as np
t = np.linspace(0,10,100)
y1 = np.sin(t)
y2 = np.cos(t)
y3 = np.sin(t) * np.cos(t)
plt.plot(t, y1, label='sin(t)')
plt.plot(t, y2, label='cos(t)')
plt.plot(t, y3, label='sin(t)*cos(t)')
plt.title('Biểu đồ các đường theo thời gian')
plt.xlabel('Thời gian (t)')
plt.ylabel('Giá trị')
plt.legend()
plt.show()
```

```python
labels = ['Q1','Q2','Q3','Q4']
A = [500,700,800,600]
B = [450,350,650,750]
C = [500,250,850,600]
#Vẽ biểu đồ thanh nhóm thể hiện doanh thu của A , B ,C trong 4 quý
import matplotlib.pyplot as plt
import numpy as np


# Tạo mảng vị trí cho các nhóm
x = np.arange(len(labels))  

# Đặt chiều rộng của các cột
width = 0.2

# Vẽ các thanh cho mỗi nhóm
fig, ax = plt.subplots(figsize=(8, 6))
bar1 = ax.bar(x - width, A, width, label='A')
bar2 = ax.bar(x, B, width, label='B')
bar3 = ax.bar(x + width, C, width, label='C')

# Thêm nhãn và tiêu đề
ax.set_xlabel('Các Quý')
ax.set_ylabel('Doanh thu')
ax.set_title('Doanh thu của A, B, và C trong các Quý')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

# Hiển thị biểu đồ
plt.tight_layout()
plt.show()
```

```python
#Biểu đồ tròn
Cty = ['A','B','C','D']
thiphan = [30,25,25,20]
color = ['#ff9999','#66b3ff','#99ff99','#ffcc99']
#Tạo biểu đồ tròn biểu hiện tỉ lệ thị phần của 4 công ty
plt.figure(figsize=(7, 7))
plt.pie(thiphan, labels=Cty, colors=color, autopct='%1.1f%%', startangle=140)

# Tiêu đề biểu đồ
plt.title('Tỉ lệ thị phần của 4 công ty')

# Hiển thị biểu đồ
plt.axis('equal')  # Đảm bảo biểu đồ tròn
plt.show()
```

```python
#Biểu đồ phân tán
x=np.random.rand(100)
y=np.random.rand(100)
sizes = np.random.rand(100)*300
colors = np.random.rand(100)
#Tạo biểu đồ phân tán của hai biến ngẫu nhiên
plt.figure(figsize=(8, 6))
plt.scatter(x, y, s=sizes, c=colors, alpha=0.5, cmap='viridis')

# Thêm tiêu đề và nhãn
plt.title('Biểu đồ phân tán của hai biến ngẫu nhiên')
plt.xlabel('Biến X')
plt.ylabel('Biến Y')

# Hiển thị biểu đồ
plt.colorbar()  # Hiển thị thanh màu
plt.show()
```

```python

import seaborn as sns
#Biểu đồ nhiệt
data = np.random.rand(10,10)
#Vẽ một heatmap thể hiện sự phân bố của dữ liệu trên ma trận

plt.figure(figsize=(8, 6))
sns.heatmap(data, annot=True, cmap='YlGnBu', fmt='.2f', cbar=True)

# Thêm tiêu đề
plt.title('Biểu đồ nhiệt thể hiện sự phân bố dữ liệu')

# Hiển thị biểu đồ
plt.show()
```
