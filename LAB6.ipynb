```python
# BÀI TẬP
# Cho y = 5x^6 + 3x^3 + 2x^1 + x + 2x + 5x^4+1
# Cho biết độ dốc của đa thức trên tại điểm nào
from sympy import symbols, diff

# Khai báo biến và hàm số
x = symbols('x')
y = 5*x**6 + 5*x**4 + 3*x**3 + 4*x + 1

# Tính đạo hàm của hàm số y theo biến x
dy_dx = diff(y, x)
print("Đạo hàm của hàm số là:", dy_dx)

# Tính độ dốc tại một điểm cụ thể (ví dụ: x = 2)
point = 2
slope_at_point = dy_dx.subs(x, point)
print(f"Độ dốc của hàm số tại x = {point} là: {slope_at_point}")
```

```python
#BTVN 1:
#Tạo một tensor có gia trị ban đầu là 2.0
#Định nghĩa hàm số và tính gradient
#y = x^3 +2x^2 +5x + 1
#hãy tính dy/dx tại giá trị của x
#Dùng phương pháp Gradient Descent với learning rate alpha = 0.1 để cập nhật giá trị x trong 10 vòng lặp


import torch

# Tạo một tensor có giá trị ban đầu là 2.0 và yêu cầu tính gradient
x = torch.tensor([2.0], requires_grad=True)

# Định nghĩa hàm số
y = x**3 + 2*x**2 + 5*x + 1

# Tính gradient
y.backward()
print(f"Giá trị của dy/dx tại x = {x.item()} là: {x.grad.item()}")

# Gradient Descent
learning_rate = 0.1
for i in range(10):
    x.grad.zero_()  # Đặt gradient về 0 trước khi tính toán lại
    y = x**3 + 2*x**2 + 5*x + 1
    y.backward()  # Tính gradient
    with torch.no_grad():
        x -= learning_rate * x.grad  # Cập nhật giá trị của x
    print(f"Vòng lặp {i+1}: x = {x.item()}, dy/dx = {x.grad.item()}")
```

Tạo một tập dữ liệu giả lập với x là số giờ học( ngẫu nhiên từ 1 - 10) và y là số điểm được tính theo công thức y =
3x + 5+ noise
Với noise là một giá trị ngẫu nhiên nhỏ
1. Khởi tạo tham số w và b ngẫu nhiên với requires _grad= true
2. tính MSE
3. Tính gradient
4. Cập nhật tham số w và b bằng gradient Descent với Learning rate alpha = 0.01
5. Lặp lại quá trình trên trong 100 vòng lặp và quan sát sự hội tụ của mô hình



```python
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np

# 1. Tạo tập dữ liệu giả lập
np.random.seed(42)
torch.manual_seed(42)

# Số giờ học (x) từ 1 đến 10
x = torch.randint(1, 11, (100, 1), dtype=torch.float32)

# Số điểm (y) theo công thức y = 3x + 5 + noise
noise = torch.randn(100, 1) * 1.5  # noise nhỏ
y = 3 * x + 5 + noise

# 2. Khởi tạo tham số w và b ngẫu nhiên với requires_grad=True
w = torch.randn(1, requires_grad=True)
b = torch.randn(1, requires_grad=True)

# 3. Định nghĩa hàm MSE (Mean Squared Error)
def mse_loss(y_pred, y_true):
    return ((y_pred - y_true) ** 2).mean()

# 4. Định nghĩa optimizer (SGD với learning rate 0.01)
optimizer = optim.SGD([w, b], lr=0.01)

# 5. Huấn luyện mô hình trong 100 vòng lặp
for epoch in range(100):
    # Dự đoán y
    y_pred = w * x + b

    # Tính loss
    loss = mse_loss(y_pred, y)

    # Tính gradient
    loss.backward()

    # Cập nhật tham số w và b bằng gradient descent
    optimizer.step()

    # Đặt gradient về 0 cho lần lặp tiếp theo
    optimizer.zero_grad()

    # In kết quả sau mỗi 10 vòng lặp
    if (epoch + 1) % 10 == 0:
        print(f'Epoch [{epoch+1}/100], Loss: {loss.item():.4f}, w: {w.item():.4f}, b: {b.item():.4f}')

# Kết quả cuối cùng
print(f'Final w: {w.item():.4f}, Final b: {b.item():.4f}')
```

```python
import torch

# Tạo tensor rỗng với kích thước (2, 3)
empty_tensor = torch.empty(2, 3)
print("Empty Tensor:\n", empty_tensor)
# Tạo tensor chứa toàn số 0 với kích thước (2, 3)
zeros_tensor = torch.zeros(2, 3)
print("Zeros Tensor:\n", zeros_tensor)
# Tạo tensor chứa toàn số 1 với kích thước (2, 3)
ones_tensor = torch.ones(2, 3)
print("Ones Tensor:\n", ones_tensor)
# Tạo tensor ngẫu nhiên với kích thước (2, 3)
random_tensor = torch.rand(2, 3)
print("Random Tensor:\n", random_tensor)
# Tạo một tensor có kích thước (2, 3)
tensor = torch.tensor([[1, 2, 3], [4, 5, 6]])

# Reshape tensor thành kích thước (3, 2)
reshaped_tensor = tensor.view(3, 2)
print("Reshaped Tensor with view:\n", reshaped_tensor)
# Tạo một tensor khác với kích thước (3, 2)
other_tensor = torch.rand(3, 2)

# Reshape tensor ban đầu thành cùng kích thước với other_tensor
reshaped_tensor_as = tensor.view_as(other_tensor)
print("Reshaped Tensor with view_as:\n", reshaped_tensor_as)
```
